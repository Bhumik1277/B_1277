```python
import numpy as np
import matplotlib.pyplot as plt
x_pos=0
y_pos=0
x_direct=1
y_direct=1
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct)
ax.set_title('Quiver plot with one arrow')
```

```python
x = np.linspace(-2, 0.55, 10)
y = np.linspace(-2, 0.55, 7)
X, Y = np.meshgrid(x, y)
U = np.cos(X)*Y
V = np.sin(Y)*X

plt.figure()
plt.title("Quiver Plot with Multiple Arrows Forming Patterns")
plt.quiver(X, Y, U, V, color='r',scale=20)

plt.show()

```
