```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
print(df.head())
```

```python
mean_age = df['age'].mean()
mean_fare = df['fare'].mean()
print(f"Mean of 'age': {mean_age}")
print(f"Mean of 'fare': {mean_fare}")
```

```python
df_cleaned = df.drop_duplicates()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['age'].dropna(), kde=True, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['fare'], kde=True, bins=30, color='green', edgecolor='black')
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='survived', data=df, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
median_age = df['age'].median()
mode_age = df['age'].mode()
print(f"Median of 'age': {median_age}")
print(f"Mode of 'age': {mode_age}")

```
