```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/bus.csv')
```

```python
print("Bus Data:")
print(df)
```

```python
mean_lat = df['lat'].mean()
mean_lon = df['lon'].mean()
print("\nMean Latitude:", mean_lat)
print("Mean Longitude:", mean_lon)
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.scatter(df['lon'], df['lat'], color='blue', label='Bus Locations')
```

```python
plt.scatter(mean_lon, mean_lat, color='red', label='Mean Location', s=100, marker='X')
```

```python
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Bus Locations and Mean Location')
plt.legend()
```

```python
plt.grid(True)
plt.show()
```
