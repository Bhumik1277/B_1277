```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/HollywoodMovies.csv')
```

```python
df_selected = df.head(20)
```

```python
numerical_columns = ['RottenTomatoes', 'AudienceScore', 'OpeningWeekend', 'BOAvgOpenWeekend',
                     'DomesticGross', 'ForeignGross', 'WorldGross', 'Budget', 'Profitability', 'OpenProfit']

means = df[numerical_columns].mean()
print("\nMean Values of Numerical Columns:")
print(means)
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.figure(figsize=(12, 8))
df_selected[['Movie', 'DomesticGross', 'ForeignGross', 'Budget']].set_index('Movie').plot(kind='bar', stacked=False, width=0.8)

```

```python
plt.title('Domestic Gross, Foreign Gross, and Budget for the First 20 Movies')
plt.ylabel('Amount (in millions)')
plt.xlabel('Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['DomesticGross'], df['ForeignGross'], color='purple', alpha=0.6)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['DomesticGross'], df['ForeignGross'], color='purple', alpha=0.6)
plt.title('Domestic Gross vs Foreign Gross')
plt.xlabel('Domestic Gross (in millions)')
plt.ylabel('Foreign Gross (in millions)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['RottenTomatoes'], bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Rotten Tomatoes Scores')
plt.xlabel('Rotten Tomatoes Score (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
